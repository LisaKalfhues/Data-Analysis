### Data analysis
##### Web scraping
##### Data set: Song lyrics scaped from www.lyrics.com

```python
import requests
import re
from collections import Counter
import time
import numpy as np
import pandas as pd
```

```python
# Copy the URLs
URL_Chvrches = 'https://www.lyrics.com/artist/Chvrches/2754697'
URL_Lizzo = 'https://www.lyrics.com/artist/Lizzo/2869948'
headers =  {'User-Agent': 'Mozilla/5.0 (X11; Linux i686; rv:2.0b10) Gecko/20100101 Firefox/4.0b10'}
```

```python
# Use the requests module to download that page
response_Chvrches = requests.get(url = URL_Chvrches, headers = headers)
response_Chvrches
response_Lizzo = requests.get(url = URL_Lizzo, headers = headers)
response_Lizzo
```

```python
# Check request status
response_Chvrches.status_code
response_Lizzo.status_code
```

```python
# Save the page to a text file
chvrches_html = response_Chvrches.text
with open('chvrches_html.txt', 'w') as file_chvrches:
    file_chvrches.write(chvrches_html)
    file_chvrches.close()
```

```python
lizzo_html = response_Lizzo.text
with open('lizzo_html.txt', 'w') as file_lizzo:
    file_lizzo.write(lizzo_html)
    file_lizzo.close()
```

```python
# Open the page in a text editor
with open('chvrches_html.txt', 'r') as file_chvrches:
    chvrches_html_read = file_chvrches.read()
    print(chvrches_html_read)
    file_chvrches.close()
```

```python
with open('lizzo_html.txt', 'r') as file_lizzo:
    lizzo_html_read = file_lizzo.read()
    print(lizzo_html_read)
    file_lizzo.close()
```
