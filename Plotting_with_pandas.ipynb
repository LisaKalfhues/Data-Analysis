### Data analysis
##### Plotting with pandas
##### Data set: Passengers of the Titanic

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = sns.load_dataset('titanic')
df
```

```python
# Create a bar plot indicating the number of first, second and third class passengers
sns.countplot(data=df, x='class',color='plum')
```

```python
# Plot a histogram of the age column
sns.histplot(data=df, x='age', color='plum')
```

```python
# Let the histogram have 20 bins
sns.histplot(data=df, bins=20, x='age', color='plum')
```

```python
# Create a boxplot of the age column
sns.boxplot(data=df, y='age', color='plum', width=0.3)
```

```python
# Create a scatterplot of fare over age
sns.scatterplot(data=df, x='age', y='fare', color='plum')
```

```python
# Create a bar plot that shows the survived vs. dead passengers with one bar group for each gender
sns.countplot(data=df, x='alive', hue='sex', color='plum')
```

```python
# Create a heatmap from the correlation matrix of all columns
sns.heatmap(df.corr())
```
